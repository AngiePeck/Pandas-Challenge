```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
csv_path = "Pymoli.csv"

# Read Purchasing File and store into Pandas data frame
pymoli_df = pd.read_csv(csv_path)
pymoli_df
```

## Player Count

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#unique items
item_count = len(pymoli_df["Item ID"].unique())

#average price of items
average_price = pymoli_df["Price"].mean()

#number of purchases
purchase_count = len(pymoli_df["Purchase ID"].unique())

#total revenue
total_revenue = pymoli_df["Price"].sum()

#summary of purchases
pymoli_summary_table = pd.DataFrame({"Number of Unique Items" : [item_count],
                                    "Average Price": [average_price],
                                    "Number of Purchases": [purchase_count],
                                    "Total Revenue": [total_revenue]})
#format numbers
pymoli_summary_table["Average Price"]=pymoli_summary_table["Average Price"].map("${:,.2f}".format)
pymoli_summary_table["Total Revenue"]=pymoli_summary_table["Total Revenue"].map("${:,.2f}".format)

pymoli_summary_table
```

## Gender Demographics



* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#make a new dataframe with duplicate users removed, this will be used in the next summary
sorted_pymoli_df = pymoli_df.sort_values("SN")
cleaned_pymoli_df = sorted_pymoli_df.drop_duplicates(subset="SN", keep="first")

#total player count
total_players = cleaned_pymoli_df["Gender"].count()
total_players
```

```python
#group the data by gender
group_gender_df = cleaned_pymoli_df.groupby(["Gender"])

#how many of each gender
gender_counts = group_gender_df["Gender"].count()
gender_counts
```

```python
gender_df = cleaned_pymoli_df[["Gender"]]

#counts for male, female, and other
male_count = gender_df.loc[gender_df["Gender"] == "Male",:]
male_count = male_count["Gender"].count()
female_count = gender_df.loc[gender_df["Gender"] == "Female",:]
female_count = female_count["Gender"].count()
other_count = gender_df.loc[gender_df["Gender"] == "Other / Non-Disclosed",:]
other_count = other_count["Gender"].count()

#percents for male, female, and other
percent_male = round((male_count/player_count)*100,2).astype("float")
percent_male = "{:.2f}%".format(percent_male)
percent_female = round((female_count/player_count)*100,2).astype("float")
percent_female = "{:.2f}%".format(percent_female)
percent_other = round((other_count/player_count)*100,2).astype("float")
percent_other = "{:.2f}%".format(percent_other)
percent_other

#summary of gender demographic
gender_demographic_df = pd.DataFrame({
                                    "Total Count":[male_count, female_count, other_count],
                                    "Percentage of Players":[percent_male, percent_female, percent_other]
                                    })
#index the table with male, female, and other
gender_demographic_df.index = ["Male", "Female", "Other/Non-Disclosed"]
gender_demographic_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins for <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40+
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
#names for bins
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
cleaned_pymoli_df["Age Groups"]=pd.cut(cleaned_pymoli_df["Age"], age_bins, labels=bin_names)
#cleaned_pymoli_df group into age bins
group_age_bins = cleaned_pymoli_df.groupby(["Age Groups"])

#count of each bin for the total count column
age_bin_count = group_age_bins["Age Groups"].count()
age_bin_count

#percentage of each age group
percent_players = (age_bin_count/total_players)*100
percent_players

#create an age demographics summary dataframe
age_demographics_df = pd.DataFrame({"Total Count":age_bin_count,
                                   "Percentage of Players":percent_players
                                   })
#format the percentage column to have percent and two decimal places
age_demographics_df["Percentage of Players"]=age_demographics_df["Percentage of Players"].map("%{:,.2f}".format)
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#use original df with all purchases!
#create bins for <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40+
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
#names for bins
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
pymoli_df["Age Groups"]=pd.cut(pymoli_df["Age"], age_bins, labels=bin_names)
#pymoli_df group into age bins
age_bins = pymoli_df.groupby(["Age Groups"])

#purchase count for each age group
purchase_count = age_bins["Purchase ID"].count()
#purchase_count

#average purchase price
average_price = age_bins["Price"].mean()
#average_price

#total purchase value
purchase_total = age_bins["Price"].sum()
#purchase_total

#average total purchase per person
ave_per_person = purchase_total/age_bin_count
#ave_per_person

#create a purchasing analysis (age) summary
purchasing_analysis = pd.DataFrame({"Purchase Count":purchase_count,
                                   "Average Purchase Price":average_price,
                                   "Total Purchase Value":purchase_total,
                                   "Avg Total Purchase per Person":ave_per_person})

#format values, add dollar signs and two decimal places to currency
purchasing_analysis["Average Purchase Price"]=purchasing_analysis["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis["Total Purchase Value"]=purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis["Avg Total Purchase per Person"]=purchasing_analysis["Avg Total Purchase per Person"].map("${:,.2f}".format)


purchasing_analysis

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
id_name_price = pymoli_df.loc[:,["Item ID", "Item Name", "Price"]]
id_name_price
group_id_name = id_name_price.groupby(["Item ID","Item Name"])

#item name


#purchase count
purchase_count = group_id_name["Item Name"].count()
purchase_count.sort_values(ascending=False)
 
#item price
item_price = group_id_name["Price"].mean().map("${:,.2f}".format)


#total purchase value
total_pur_value = group_id_name["Price"].sum()
#total_pur_value.sort_values(ascending=False)

#summary of most popular items
most_pop_unsorted = pd.DataFrame({"Purchase Count": purchase_count,
                        "Item Price": item_price,
                        "Total Purchase Value": total_pur_value 
                        })
most_pop_sorted = most_pop_unsorted.sort_values("Purchase Count",ascending=False).head()
most_pop_sorted["Total Purchase Value"]=most_pop_sorted["Total Purchase Value"].map("${:,.2f}".format)
most_pop_sorted
```
